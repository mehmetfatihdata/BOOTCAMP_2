```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
prices = pd.read_csv('dataset/house_prices.csv', index_col=0)
prices.tail(2)
```

## Q1 :
Ev fiyatları modelinizi tekrar çalıştırın ve F-testi, R-kare, ayarlanmış R-kare, AIC ve BIC kullanarak modelinizin uygunluğunu değerlendirin.

```python
high=prices.corr().SalePrice.abs().sort_values(ascending=False).index[0:11]
```

```python
Y = prices['SalePrice']
X = prices[['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea',
       'TotalBsmtSF', '1stFlrSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt',
       'YearRemodAdd']]
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

""""R-squared:	0.774,
Adj. R-squared:	0.772"""

## Q3:
Modelinizin uygunluğunu iyileştirmek için, bazı değişkenleri ekleyerek veya kaldırarak farklı model özelliklerini deneyin.

```python
Y = prices['SalePrice']
X = prices[['OverallQual', 'GrLivArea', 'GarageCars',
       'TotalBsmtSF', '1stFlrSF', 'FullBath', 'YearBuilt',
       'YearRemodAdd']]
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

Ayrıca p değeri uygun olmayan değişkenler modelden çıkartılabilir. Zaten etkisiz olan iki değişkenin çıkması tahmin performansını değiştirmeyecektir. Ancak BIC değerindeki düşüşe dikkat etmek gerekir. Bu nedenle istatistiksel olarak anlamsız olan değişkenler modelden çıkarılmadılır. 

```python
Y = prices['SalePrice']
prices['Garage']=prices['GarageCars']*prices['GarageArea']
X = prices[['OverallQual', 'GrLivArea', 'Garage',
       'TotalBsmtSF', '1stFlrSF', 'FullBath', 'YearBuilt',
       'YearRemodAdd']]
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```
